# Some fancy plots 

```python
import matplotlib.pyplot as plt
import numpy as np
import pathlib

im_dir = pathlib.Path('../../datasets/fLoc_stimuli/')
```

```python
%matplotlib inline
```

```python
files = sorted(list(im_dir.glob('*jpg')))
```

```python
files_toshow = np.random.choice(files, replace=False, size=(12,))
```

```python
fig, axs = plt.subplots(3, 4, figsize=(10,7.5))

for f, ax in zip(files_toshow, axs.flatten()):
    n = files.index(f)
    ax.imshow(plt.imread(f), cmap='gray')
    ax.set_title(f'Image {n}', fontsize=14)
    ax.axis('off')
plt.tight_layout()
```
